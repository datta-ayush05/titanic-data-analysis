## Step 1 : Load and Preview the Dataset
We'll begin by importing the dataset and taking a first look at it's structure and contents

## Step 4 : Bivariate Analysis
Here, we analyze the dependence and relationship between two variables
- Analysis and dependence of survival rate on factors such as gender, passenger class, age groups, embarkement port, etc.
- Graphical representation and comparison of the survival rates with change in values of other variables (or factors)

```python
# Sex

# Survival rate with gender
print((df.groupby('Sex')['Survived'].mean()*100).round(2))

#Distribution of survivors per gender
print((((df.groupby('Sex')['Survived'].sum()) / df['Survived'].sum()) * 100).round(2))
```

```python
# Age Group

# Defining age groups
def f(x) :
    if x <= 12 :
        return 'Children'
    elif x > 12 and x <= 19 :
        return 'Teens'
    elif x > 19 and x <= 35 :
        return 'Young Adults'
    elif x > 35 and x <= 55 :
        return 'Middle Aged'
    else :
        return 'Seniors'
df['AgeGroup'] = df['Age'].map(f)

# Survival rate with age groups
print(((df.groupby('AgeGroup')['Survived'].mean()) * 100).round(2).sort_values(ascending = False))

# Distribution of survivors with age groups
print((((df.groupby('AgeGroup')['Survived'].sum()) / df['Survived'].sum()) * 100).round(2).sort_values(ascending = False))
```

```python
# Survival rate per class
print((df.groupby('Pclass')['Survived'].mean()*100).round(2))

# Distribution of survivors per class
print((((df.groupby('Pclass')['Survived'].sum()) / df['Survived'].sum()) * 100).round(2))
```

```python
# Survival rate per port
print((df.groupby('Embarked')['Survived'].mean() * 100).round(2))

# Distribution of survivors per port
print((((df.groupby('Embarked')['Survived'].sum()) / df['Survived'].sum()) * 100).round(2))
```

```python
df['HasCabin'] = df['Cabin'].notna().astype(int)

# Survival rate vs Cabin availability
print((df.groupby('HasCabin')['Survived'].mean() * 100).round(2))

# Distribution of survivors vs Cabin availability
print(((df.groupby('HasCabin')['Survived'].sum() / df['Survived'].sum()) * 100).round(2))
```

```python
def f(x) :
    if x < 100 :
        return 'Tier 1'
    elif x >= 100 and x < 200 :
        return 'Tier 2'
    elif x >= 200 and x < 300 :
        return 'Tier 3'
    elif x >= 300 and x < 400 :
        return 'Tier 4'
    elif x >= 400 and x < 500 :
        return 'Tier 5'
    else :
        return 'Tier 6'
df['FareCategory'] = df['Fare'].map(f)

# Survival rate per fare category
print((df.groupby('FareCategory')['Survived'].mean() * 100).round(2))

# Distribution of survivors per fare category
print(((df.groupby('FareCategory')['Survived'].sum() / df['Survived'].sum()) * 100).round(2))
```

### Bivariate Plots
Let's visualize the survival rates against all these factors through plots (barplots, countplots). We'll do
- Survival rate vs Sex
- Survival rate vs Age groups
- Survival rate vs Pclass
- Survival rate vs Embarkation port
- Survival rate vs Cabin availability
- Survival rate vs Fare category

```python
fig, ax = plt.subplots(figsize = (4, 4))
df.groupby('Sex')['Survived'].mean().plot.bar(ax = ax, color = ['teal', 'blue'])
ax.set(title = 'Survival Rate by Sex', ylabel = 'Survival rate', ylim = (0, 1))
ax.grid()
plt.show()
```

```python
age_order = ['Children', 'Teens', 'Young Adults', 'Middle Aged', 'Seniors']
fig, ax = plt.subplots(figsize = (6,4))
df.groupby('AgeGroup')['Survived'].mean().reindex(age_order).plot(ax = ax, color = 'teal', marker = 'o')
ax.set(title = 'Survival Rate by Age Groups', ylabel = 'Survival Rate', ylim = (0,1))
ax.grid()
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (6,4))
df.groupby('Pclass')['Survived'].mean().plot.bar(ax = ax, color = ['teal', 'blue', 'indigo'])
ax.set(title = 'Survival Rate by Passenger class', ylabel = 'Survival Rate', ylim = (0,1))
ax.set_xticklabels(['1', '2', '3'], rotation = 0)
ax.grid()
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (6, 4))
df.groupby('Embarked')['Survived'].mean().plot.bar(ax = ax, color = ['seagreen', 'forestgreen', 'limegreen'])
ax.set(title = 'Survival Rate by Embarkment Port', ylabel = 'Survival rate', ylim = (0, 1))
ax.set_xticklabels(['C', 'Q', 'S'], rotation = 0)
ax.grid()
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (4, 4))
df['Survived'].groupby(df['Cabin'].isna()).mean().plot.bar(ax = ax, color = ['teal', 'blue'])
ax.set(title = 'Survival Rate by Cabin Availability', xlabel = 'Cabin Availability', ylabel = 'Survival rate', ylim = (0, 1))
ax.set_xticklabels(['Does not have Cabin', 'Has Cabin'], rotation = 0)
ax.grid()
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (6,4))
df.groupby('FareCategory')['Survived'].mean().plot(ax = ax, color = 'seagreen', marker = 'o')
ax.set(title = 'Survival Rate by Fare Category', ylabel = 'Survival Rate', ylim = (0,1))
ax.grid()
plt.show()
```

### Bivariate Analysis Observations
In this section, we analyzed how Survival Rates vary with key features. For each factor, we plotted Survival Rate vs Factor to understand their individual influence on survival outcomes.
Key Observations :
- Sex : Female passengers had a much higher survival rate than males
- Pclass : Passengers in 1st class survived more often than those in 2nd and 3rd classes
- Port of Embarkation : Port C had the highest survival rate followed by ports Q and S
- Age Group : Younger age groups, especially children, had higher survival chances
- Fare Group : Passengers in higher fare brackets generally survived more
- Cabin Availability : Those with cabin info (non-null values) had better survival odds - likely correlated with class
For each of these factors, we also investigated the distribution of survivors
